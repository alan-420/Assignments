## Notebook Imports

```python
from sklearn import decomposition  
from sklearn import datasets   
import pandas as pd
import plotly_express as px
```

## Creating the dataframe

```python
iris_data = datasets.load_iris().data
iris_target = datasets.load_iris().target
iris_cols = datasets.load_iris().feature_names
target_names = datasets.load_iris().target_names
```

```python
iris = pd.DataFrame(iris_data, columns= iris_cols)
iris['target']= iris_target
```

```python
iris['target'] = iris.target.map({0:'setosa', 1:'versicolor', 2: 'virginica'})
```

```python
iris.head()
```

## Decomposition of data using Principal Component Analysis.

```python
decom = decomposition.PCA(n_components=3)
decomposed_data = decom.fit_transform(iris_data)
```

```python
new_df = pd.DataFrame(decomposed_data, columns=['A','B','C'])
new_df['Species'] = iris.target
```

## Plotting the data

```python
px.scatter_3d(new_df, x = 'A',y = 'B',z = 'C',color="Species")
```
